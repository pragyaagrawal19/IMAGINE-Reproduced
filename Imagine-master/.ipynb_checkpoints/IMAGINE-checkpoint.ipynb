# Reproducing results from imagine

This notebook is a walk-through the *IMAGINE* learning algorithm, from the [Language as a Cognitive Tool to Imagine Goals in Curiosity Driven Exploration](https://arxiv.org/pdf/2002.09253.pdf) paper. 

This implementation is a representation of how artificial agents can learn open-ended andcreative learning skills, which are generally commonly seen in childrenâ€™s.  It is a deep re-inforcement learning (RL) model that uses compositionality of language as mechanism tovisualize the new interesting results they never experienced before.  For providing languagedescriptions to agents, this model provides a social peer (SP).This implementation of IMAGINE ( Intrinsic Motivation And Goal Invention For Explo-ration)  focuses on out-of-distribution goal generation.  This model is language condi-tioned in the sense that: it is never given any instruction or reward but it sets its own goal andlearn its own internal reward function.  The lingual foundation to goal imagination is basedon construction grammar (CG). It describes language acquisition as a route that starts withpattern imitation and equivalence class discovery argument swapping, before going towards recognition and composition of more abstract patterns. It uses compositionality of langauge to generate new goals from known training goals.

#### <strong> 1.1.2 Second Role of the Social Partner: Providing Descriptions</strong>



The second role of the social partner is to provide language descriptions about the goals reached by the agent __at the end of an episode__.

```python
def generate_animated_rollout(env, goal_str, initial_o, policy, policy_language_model, reset=False):
    # In order to build vizualization
    fig = plt.figure()
    ims = []

    ## Rollout of a policy
    if reset:
        env.reset()
        initial_o = env.unwrapped.reset_with_goal(goal_str)
        o = initial_o.copy()
    else:
        o = initial_o.copy()
    goal_encoding = policy_language_model.encode(goal_str)
    episode={'obs':[]}
    for t in range(30):
        action = policy.get_actions(o, goal_encoding)
        o,_,_,_ = env.step(action)
        episode['obs'].append(o)
        env.render(close=True)
        obs = pygame.surfarray.array3d(env.viewer).transpose([1,0,2])
        plt.rcParams['figure.figsize'] = [12, 8]
        im = plt.imshow(obs, animated=True)
        ims.append([im])
      
    ani = animation.ArtistAnimation(fig, ims, interval=400, blit=False,
                                  repeat=False)
    return ani, episode, o, obs
```

```python
%%capture
ani, episode, o, obs = generate_animated_rollout(env, goal_str, initial_o, policy, policy_language_model)
```

### <strong> 1.2 The Reward Function Module </strong>

------------

#### <strong> 1.2.1 The architecture explained </strong> 

The  Reward  Function  is  implemented  using  a  Deep  Set [Deep Set](https://arxiv.org/abs/1703.06114) architecture.   Deep  set  isnetwork architecture that is based on permutation in-variance properties. Each input mappedseparately to a latent space by using a shared network, then these representations passed topermutation-invariant function like sum or mean to make sure permutation in-variance of thewhole function.Object are grouped into object dependent sub object reward function, each one has objectlevel reward by a same Neural network (N Nreward).  It can be thought as a single rewardfunction which estimates whether goal is achieved by object or not with some probabilitye.g. >0.5.  After that these rewards are mapped to global binary reward using a logical orfunction. It state that: if any of object verifies the goal then, whole-scene also verifies it.The computation of reward function also use gated-attention mechanism. Though thismechanism goal embedding is cast into an attention vector and then it is combined to objectdependent state though term by term product to formN Nreward.  (Scaling object specificfeature based on interpretation of goal embedding.)

The small diagram below gives an illustration of the reward function architecture:

<img src="https://drive.google.com/uc?export=view&id=1-74Gx9PkoPz4DC2gNPkdlSDeQsRVCq5G" alt="drawing" width="850"/>

Representation of rewards acheived

```python
def compute_reward_probabilities(state, goal_str, reward_language_model, reward_function, env):
    goal_encoding = reward_language_model.encode(goal_str)
    input_o = torch.tensor(state).float().view(1,len(o))
    input_goal = torch.tensor(goal_encoding).float().view(1,len(goal_encoding))
    reward_per_object = reward_function.reward_function.compute_logits_before_or(input_o, input_goal)
    reward_per_object = [elem.detach()[0][0].item() for elem in reward_per_object]
    objects = [obj.object_descr['colors'] + ' ' + obj.object_descr['types'] for obj in env.objects]
    d = {'Objects':objects, 'Reward':reward_per_object}
    df_rew = pd.DataFrame(d)
    return df_rew

def generate_other_possible_goal(goal_str, env):
    objects = [obj.object_descr['colors'] + ' ' + obj.object_descr['types'] for obj in env.objects]
    other_objects = [o for o in objects if o not in goal_str]
    return ' '.join(random.sample(['Grow'], 1) + random.sample(other_objects, 1))
```

```python
df_rew1 = compute_reward_probabilities(o, goal_str, reward_language_model, reward_function, env)
df_rew2 = compute_reward_probabilities(o, other_goal_str, reward_language_model, reward_function, env)

f, axes = plt.subplots(1, 2, figsize=(16, 7),sharey=True)
sns.despine(left=True)
sns.barplot(x="Objects", y="Reward", data=df_rew1, ax=axes[0]).set_title(goal_str)
for index, row in df_rew1.iterrows():
    axes[0].text(row.name, row.Reward + 0.03, '{:.2f}'.format(row.Reward), color='black', ha="center")
axes[0].plot([-0.5, 2.6], [0.5, 0.5], linestyle='--', color='k')

sns.despine(left=True)
sns.barplot(x="Objects", y="Reward", data=df_rew2, ax=axes[1]).set_title(other_goal_str)
for index, row in df_rew2.iterrows():
    axes[1].text(row.name, row.Reward + 0.03, '{:.2f}'.format(row.Reward), color='black', ha="center")
axes[1].plot([-0.5, 2.6], [0.5, 0.5], linestyle='--', color='k')

```

deepset architecture operating at the object level is confidently detecting the goal achievement (left): $R_{object}>0.5$ (above dashed line).

For the goal that is not fulfilled (right), the architecture correctly predicts that there is no reward for the three objects. However, we can notice that the model outputs a higher signal for the object mentioned in the goal description. This means that the reward function is able to understand which object is used in the goal description without outputting a reward signal > 0.5 if the goal involving this object is not achieved.

# Comparision of IMAGINE with Simple conjuction based heuristic (SCBH)

### <strong> 1.4 The Goal Imagination Module vs random sentence generation module </strong>

Goal imagination module in IMAGINE paper was inspired by Construction Grammar. For each pair of known goals that differ by exactly one word, the two differing words are considered equivalent. This means they can be exchanged in any known goal to form a novel goal. 
In SCBH goals were generated randomly by taking objects and actions randomly from the enviroment.(It was ensured that the goals was not from training set). If we compare the coverage of Imagine with this method, randomly generated goals were able to give almost simillar coverage as IMAGINE, also due to randomness of method the new descriptions that would be generated will be much more diverse (not much of goals will be in testing set). So we can hence observe that thought of language compositionality in language (proposed by IMAGINE) was not seems to be reasonable. 

Comparision of coverage(percentage of  Sentences in the test set, found by generation) and precision(percentage of new description found in test set) of CGH and SCBH

```python
    from src.imagine.goal_generator.simple_sentence_generator import SentenceGeneratorHeuristic, simple_conjuction_based_heuristic
    from src.playground_env.env_params import get_env_params
    from src.playground_env.descriptions import generate_all_descriptions
    # np.random.shuffle(train_descriptions)
    env_params = get_env_params()
    train_descriptions, test_descriptions, _ = generate_all_descriptions(env_params)
    import numpy as np
    print("description for CGH method")
    generator1 = SentenceGeneratorHeuristic(train_descriptions, test_descriptions, None, method="CGH")
    generator1.update_model(train_descriptions)
    new_descriptions1 = generator1.generate_sentences()

    p_found_in_test = sum([d in test_descriptions for d in new_descriptions1]) / len(test_descriptions)
    p_not_in_test = sum([d not in test_descriptions for d in new_descriptions1]) / len(new_descriptions1)
    p_in_test = sum([d in test_descriptions for d in new_descriptions1]) / len(new_descriptions1)
    print('Percentage of the test set found (coverage):', p_found_in_test)
    print('Percentage of the new descriptions that are not in the test', p_not_in_test)
    print('Percentage of the new descriptions that are in the test set (precision)', p_in_test)

    print('\n Sentences in the generated set, not in the test: \n', set(new_descriptions1) - set(test_descriptions))
    print('\n Sentences in the test set, found by generation: \n',
          set(new_descriptions1).intersection(set(test_descriptions)))
    print('\n Sentences in the test set, not in the generated set: \n', set(test_descriptions) - set(new_descriptions1))
    print("new_description_legnth",len(new_descriptions1))
    print("""""""""""""""""""""""""""""""""""""""""""""""""")
    print("""""""""""""""""""""""""""""""""""""""""""""""""")
    print('***************************************************')
    print("description for SCBH method")
    generator = simple_conjuction_based_heuristic(train_descriptions, test_descriptions, None, method="SCBH")
    new_descriptions = generator.generate_sentences()

    p_found_in_test = sum([d in test_descriptions for d in new_descriptions]) / len(test_descriptions)
    p_not_in_test = sum([d not in test_descriptions for d in new_descriptions]) / len(new_descriptions)
    p_in_test = sum([d in test_descriptions for d in new_descriptions]) / len(new_descriptions)
    print('Percentage of the test set found (coverage):', p_found_in_test)
    print('Percentage of the new descriptions that are not in the test ', p_not_in_test)
    print('Percentage of the new descriptions that are in the test set (precision)', p_in_test)

    print('\n Sentences in the generated set, not in the test: \n', set(new_descriptions) - set(test_descriptions))
    print('\n Sentences in the test set, found by generation: \n',
          set(new_descriptions).intersection(set(test_descriptions)))
    print('\n Sentences in the test set, not in the generated set: \n', set(test_descriptions) - set(new_descriptions))
    print(len(new_descriptions))
```

# comparison of Average success rate (over 30 episods)

##### <strong> Comaprison of CGH and SGBH </strong>

```python
import json
import matplotlib.pyplot as plt
import numpy as np
import pickle
import os
import pandas as pd
import matplotlib

folder_paths = [ 'Reproduced/', '3633/']
name=["CGH", "SCBH"]
for path in range(len(folder_paths)):
    with open(os.path.join(folder_paths[path], 'params.json')) as json_file:
        params = json.load(json_file)

    instructions = params['train_descriptions']

    nb_instr = len(instructions)
    n_cycles = params['experiment_params']['n_cycles']
    rollout_batch_size = params['experiment_params']['rollout_batch_size']
    n_cpu = params['experiment_params']['n_cpus']
    data = pd.read_csv(os.path.join(folder_paths[path], 'progress.csv'))

    n_points = data['eval/success_goal_0'].shape[0]
    episodes = data['episode']

    n_epochs = len(episodes)
    n_eps = 1 * rollout_batch_size * n_cycles
    episodes = np.arange(n_eps, n_epochs * n_eps + 1, n_eps)
    episodes = episodes / 1000

    task_success_rates = np.zeros([n_points, nb_instr])
    goals_reached = np.zeros([n_points, nb_instr])
    for i in range(nb_instr):
        task_success_rates[:, i] = data['eval/success_goal_' + str(i)]
    zero_success_rates = task_success_rates.copy()
    for i in range(zero_success_rates.shape[0]):
        for j in range(zero_success_rates.shape[1]):
            if np.isnan(zero_success_rates[i, j]):
                zero_success_rates[i, j] = 0
    # plot success_rate for both CGH and SCBH

   
    p = plt.plot(episodes, zero_success_rates.mean(axis=1), label= name[path], linewidth=10), # , c=colors[i])
    # leg = plt.legend(['task ' + str(i) for i in range(nb_instr)], frameon=False)
    lab = plt.xlabel('Episodes (x$10^3$)')
    plt.ylim([-0.01, 1.01])
    plt.yticks([0.25, 0.50, 0.75, 1])
    lab2 = plt.ylabel('Average success rate')
plt.legend() 
plt.show()
```
